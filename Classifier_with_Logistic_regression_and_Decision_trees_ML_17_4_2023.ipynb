```python
#iris
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression #statistical approach
from sklearn.tree import DecisionTreeClassifier#classical approach
from sklearn.tree import export_graphviz
import graphviz
```

```python
iris=load_iris()
iris
```

```python
#shorter way to access list
x=[1,2,3,4]
x[:3]
x[2:]#goes to position two and print everything after that
```

```python
x=iris.data[:,:2]#read up to the second column
y=iris.target
```

```python
from sklearn.model_selection import train_test_split as ts

xtrain,xtest, ytrain,ytest= ts(x,y,test_size=0.3,random_state=42)
```

```python
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
```

```python
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

visualize logreg

```python
import matplotlib.pyplot as plt
```

```python
#above creates a grid
xx,yy=np.meshgrid(

    np.arange(x[:, 0].min() - 0.5, x[:, 0].max() +0.5, 0.02),
    np.arange(x[:,1].min()-0.5,x[:,1].max() +0.5, 0.02),
)
pred=logreg.predict(np.c_[xx.ravel(),yy.ravel()])
pred= pred.reshape(xx.shape)

plt.figure()
plt.contour(xx,yy,pred,alpha=0.8,cmap=plt.cm.RdYlBu)
plt.scatter(x[:,0], x[:,1],c=y,cmap=plt.cm.RdYlBu,edgecolor='black')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Stat approach")
plt.ylabel("Y")
plt.show
```

```python
dt_data=export_graphviz(
    dt,out_file=None, feature_names=iris.feature_names[:2],class_names=iris.target_names,
    filled=True, special_characters=True
)
graph= graphviz.Source(dt_data)
graph.render("iris")
graph
```
