```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.5, 1, 2, 2.5, 3.5, 4, 5, 5.5, 6])
y = np.array([1.5, 2, 1, 2.5, 2, 3.5, 2.5, 3, 4])

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
# m = (sum((x - x_mean) * (y - y_mean)))/(sum((x - x_mean) ** 2))

x_mean = round(np.mean(x), 4)
y_mean = round(np.mean(y), 4)

print(x_mean, y_mean)
```

## OLMSM Ordinary Least Mean Square Method



```python
m = (np.sum((x - x_mean) * (y - y_mean)))/(np.sum((x - x_mean) ** 2))

m = round(m, 4)

print(m)

# c = y_mean - m * x_mean

c = round((y_mean - m * x_mean), 4)

print(c)
```

LinearRegression
</br>
y=mx+c</br>

```python
y_pred = m * x + c

for i in range(len(x)):
    print("Actual (y): ", y[i], "Predicited (y_pred): " ,y_pred[i], "Diff: ", y[i] - y_pred)
```

```python
slope = 0
intercept = 0


learn_rate = 0.01
iterations = 20
```

```python
for i in range(iterations):
    y_pred = slope * x + intercept
    cost = np.mean((y_pred - y) ** 2)

    print("Iteration ", i + 1, " | MSE: ", round(cost, 2), " | Slope: ", round(slope, 2), " | Intercept: ", round(intercept, 2))
    print("------------" * 20)

    slope = slope + learn_rate
    intercept = intercept + learn_rate
```
